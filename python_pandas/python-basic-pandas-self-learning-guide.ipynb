## .query()
The `.query()` method is used to filter a DataFrame based on a query expression. It provides a convenient way to select a subset of data that meets specific conditions without using traditional boolean indexing.

````python
data = {'Name': ['John', 'Alice', 'Bob'],
        'Age': [25, 30, 22],
        'Salary': [50000.0, 60000.0, 45000.0],
        'IsStudent': [False, False, True]}

d1 = pd.DataFrame(data)

# Use .query() to filter rows where Age is greater than 25
result = d1.query('Age > 25')

result
````

## Dealing with Duplicate Data - .drop_duplicates()

````python
# netflix dataset has no duplicate data.
# So lets add a duplicate row to existing dataframe.
d1 = df3.copy() 
new_row = d1.loc[8806] # extract data of index = 8806
n = pd.DataFrame(new_row).T # create dataframe and transpose the result
d1 = pd.concat([d1,n],ignore_index = True) 
# The ignore_index=True parameter is used to reindex the resulting DataFrame. 
d1.tail()
````

````python
# lets remove it
dropped_df = d1.drop_duplicates(subset = ['show_id'])
dropped_df.tail()
````

**Understanding subset parameter in .drop_duplicates()**

````python
# what if show id is same but different type is there? so we need to search keys in pairs.
# lets add two new rows with same showid as s8807 and different type in one row.
d1 = df3.copy() 
new_row = d1.loc[[8806,8806]]
n = pd.DataFrame(new_row)
d1 = pd.concat([d1,n],ignore_index = True)
d1.loc[8808,'type'] = 'Drama'
d1.tail(5)
````

Now we have 3 rows with showid = s8807, in which two rows have duplicate values in all column.

````python
# let's delete the duplicate pairs.
d1.drop_duplicates(subset = ['show_id','type'])
d1.tail()
````

From above result we could see there are two rows with same `showid = s8807` but different type.

This is how subset works. We can add as many as column names.

## pd.merge()
A function used to merge two DataFrames based on common columns or indices.

It supports various types of joins, including inner, outer, left, and right joins.

````python
x1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value': [1, 2, 3]})
x2 = pd.DataFrame({'key': ['B', 'C', 'D'], 'value': [4, 5, 6]})

result = pd.merge(x1, x2, on='key', how='outer')
result
````

The `left_on` and `right_on` parameters in the `pd.merge()` function are used when you want to merge DataFrames based on columns with different names.

````python
# Sample DataFrames
left_df = pd.DataFrame({'left_key': ['A', 'B', 'C'], 'value': [1, 2, 3]})
right_df = pd.DataFrame({'right_key': ['B', 'C', 'D'], 'value': [4, 5, 6]})

# Merge based on different column names using left_on and right_on
merged_df = pd.merge(left_df, right_df, left_on='left_key', right_on='right_key', how='inner')

merged_df
````

## Create calculated fields
calculated fields refer to creating new columns in a DataFrame based on calculations involving existing columns.

Let's create a column to categories Age group as Adult,Senior Adult,Super Senior.



````python
d1 = df2.copy()
conditions = [
    (d1['Age'] >= 21) & (d1['Age'] < 40),
    (d1['Age'] >= 40) & (d1['Age'] < 60),
    (d1['Age'] >= 60)
]
choices = ['Adults', 'Senior Adults', 'Super Senior']
d1['Age_Group'] = np.select(conditions, choices, default='Unknown')
d1.tail()
````

## Aggregating data with .groupby()
`.groupby()` function is used for grouping data based on some criteria and then applying a function or summary method to each group independently.

Any summary method can be used alongside `.groupby()`, including `.min()`, `.max()`, `.mean()`, `.median()`, `.sum()`, `.mode()`, and more.

````python
# Aggregating data by one column
d1.groupby('Age_Group').median()
````

````python
# Aggregating data by multiple column
d1.groupby(['Age_Group','Outcome']).mean()
````

Additionally, you can use the `.agg()` method to apply multiple aggregation functions at once.

````python
# Applying multiple method for all columns.
d1.groupby('Age_Group').agg(['sum','mean'])
````

````python
# Applying multiple method for selected columns
d1.groupby('Age_Group').agg({'Glucose':['mean','median'],'BloodPressure':['mean','median']})
````

 It is used for binning or grouping continuous data into discrete intervals or bins.

````python
# Generate sample data with ages between 21 and 80
np.random.seed(42)
ages = np.random.randint(21, 81, size=100)

# Define bin edges
bins = [20, 30, 40, 50, 60, 70, 80]

len(ages)
````

````python
# Use pd.cut() to create age groups
age_groups = pd.cut(ages, bins=bins, right=True, labels=['21-30', '31-40', '41-50', '51-60', '61-70', '71-80'])

# Create a DataFrame for visualization
df = pd.DataFrame({'Age': ages, 'Age Group': age_groups})

# Display the DataFrame
print(df.head(10))
````

## Series.map(func)
The map() method in pandas is used to transform values in a Series based on a mapping or a provided function. 

````python
# Sample Series
s = pd.Series([1, 2, 3, 4])
s
````

````python
# Define a function for mapping
def square(x):
    return x ** 2

# Use map to transform values based on the function
squared_s = s.map(square)

print(squared_s)
````

## DataFrame.applymap(func)
 This method is used to apply a function to each element of a DataFrame. 

````python
# Sample DataFrame
data = {'A': [1.234, 2.567, 3.789],
        'B': [4.321, 5.432, 6.543]}
df = pd.DataFrame(data)
df
````

````python
# Define a custom rounding function
def custom_round(value):
    return round(value, 2)

# Apply the custom rounding function to each element of the DataFrame
rounded_df = df.applymap(custom_round)

rounded_df
````

## <font color='red'>NOTES TO REMEMBER</font>
**1. Difference between `df.columnname` and `df['columnname']`:**
* `df.columnname` is designed to retrieve a single column from a Pandas DataFrame.
* `df['columnname']` is used to retrieve single column as well as multiple columns as `df[['column1','column2']]`

**2. Understanding output of `mode()`:**
*  the mode() function returns a DataFrame with all modal values.
* If a column had a single mode, the corresponding entry in the result would be that mode.
* If there were no modes, the entry would be `NaN`.
* If there are multiple modes they are also included in the result.

**3. `.value_counts(sort,normalize)`:**
* The default sorting is based on the counts in descending order `sort = True`
* `sort = False` will sort by ascending order
* Adding the `normalize = True` argument returns relative frequencies instead of absolute counts.

**4. `pd.pivot_table()` parameters:**
* Here's a breakdown of the commonly used parameters:
    * **data** (DataFrame: The data to be used for creating the pivot table.)
    * **values=None** (Column name or list of column names to aggregate. These will be the values in the pivot table.)
    * **index=None** (Column name or list of column names to be used as the index (rows) in the pivot table.)
    * **columns=None** (Column name or list of column names to be used as the columns in the pivot table.)
    * **aggfunc='mean'** (Aggregation function or a dictionary of column names and aggregation functions.)
    * **fill_value=None** (A scalar value to replace missing values in the pivot table.)
    * **margins=False** (If True, adds an 'All' row and column to the pivot table with the overall mean/sum of the data.)
    * **margins_name='All'** (The name to use for the 'All' row and column if margins is set to True.)
    
 **5. Difference between `df.sort_values()` and `df.nlargest()`:**
 * `sort_values()` is used to sort a DataFrame or Series based on the values in one or more columns.
     * DataFrame.sort_values(by, axis=0, ascending=True, inplace=False, ...)
 * `nlargest()` is specifically used to get the first n rows with the largest values in one or more columns.
     * DataFrame.nlargest(n, columns, keep='first')
