## **CREATING VARIABLES**
R departs from the traditional approach of explicitly declaring variables. Instead, it adheres to a dynamic typing system. This means variables are implicitly created at the moment you assign a value to them using the assignment operator `<-` else `=`. The data type of the variable is then determined by the value you assign.

For instance, if you assign the number 42 to a variable named age, age becomes an integer variable.

### **🔴 Attention peers!**

the equal sign (=) can technically be used to assign values to variables in R. However, there are strong reasons why it's generally recommended to use the <- operator instead. Here's a breakdown:

- Clarity:  The <- symbol explicitly indicates assignment. It makes your code more readable, especially for those familiar with other programming languages where = is the standard assignment operator.

- Consistency:  Using <- throughout your R code promotes consistency and avoids potential confusion.  Since = can also be used for comparisons in R (e.g., x = 5 checks if x equals 5), mixing the two can lead to readability issues.

- Community Standard:  The R community overwhelmingly favors <- for variable assignment. This makes code sharing and collaboration easier as everyone understands the intended meaning.

- While the equal sign might work in some cases, it's a better practice to stick with <- to ensure clarity, consistency, and adherence to the R community standard.

```r
var1 = "Hi R Learners!"
var1
```

```r
var2 <- "Hi R Learners!!"
var2
```

## **R DATATYPES**

   These are the fundamental building blocks for storing different kinds of information.  R offers a variety of data types like:

* **Numeric**: Numbers (whole and decimals) - `10`, `3.14`, `-22`
* **Integer**: Whole numbers specifically - `10L`, `55` (optional "L" denotes integer)
* **Character** (String): Text data - `"Hello!"`, `"This is a sentence"` (enclosed in quotes)
* **Logical** (Boolean): True or False values - `TRUE`, `FALSE` (useful for decision making)
* **Complex**: Numbers with imaginary unit ("i") - `2 + 3i` (advanced math)
* **Raw**: Raw bytes of data (for specific data manipulation)

## **R OPERATORS**

**1. Arithmetic Operators: These perform basic mathematical calculations on numeric data.**

* +: Addition (e.g., x + 5)
* -: Subtraction (e.g., y - 2)
* *: Multiplication (e.g., z * 3)
* /: Division (e.g., a / b)
* ^: Exponentiation (e.g., 2 ^ 3)

**2. Relational Operators: These compare values and return logical (TRUE or FALSE) results.**

* ==: Equal to (e.g., x == 10)
* !=: Not equal to (e.g., y != "apple")
* <: Less than (e.g., z < 5)
* \>: Greater than (e.g., a > b)
* <=: Less than or equal to (e.g., x <= 10)
* \>=: Greater than or equal to (e.g., y >= "apple")

**3. Logical Operators: These combine logical values (TRUE or FALSE) to create more complex conditions**.

* &: AND (e.g., x > 0 & y < 10) - Both conditions must be true.
* |: OR (e.g., x == 5 | y == "hello") - At least one condition must be true.
* !: NOT (e.g., !(x < 0)) - Reverses the logical value.

**4. Assignment Operators: These assign values to variables. The most common is:**
* <-: Assigns a value to a variable (e.g., x <- 10).

**5. Miscellaneous Operators: R offers additional operators for specific tasks:**
* :: Sequence creation (e.g., 1:10)
* %%: Modulo (remainder after division)
Remember, the order of operations (precedence) matters in R expressions. Use parentheses to control the order of evaluation if needed.

```r
n2 <- n1
n2$director[is.na(n2$director)] <- "Unknown"
colSums(is.na(n2))
```

```r
n2 <- n1
print("Before imputation")
colSums(is.na(n2))

n2 <- n2 %>%
mutate(director = ifelse(is.na(director),"Unknown",director))

print("after imputation")
colSums(is.na(n2))

head(n2,10)
```

## RANK / DENSE_RANK / ROWNUMBER

`row_number()` is typically used to create a sequence of row numbers.

`rank()` is used to rank values, and tied values can result in non-sequential ranks. where tied values may receive different ranks, and there may be gaps between ranks.

`dense_rank()` produce ranks which are consecutive integers with no gaps and tied values receive the same rank.

```r
d1 %>% group_by(gender) %>%
mutate(rownumber = row_number(),BG_rank = rank(blood_glucose_level,ties.method = "min"),BG_d_rank = dense_rank(blood_glucose_level)) %>%  select(c(gender,blood_glucose_level,BG_d_rank,BG_rank,rownumber))
```

* **%b**: Abbreviated month name in the current locale (e.g., "Jan").
* **%B**: Full month name in the current locale (e.g., "January").
* **%y**: Year without century (00-99). Values 00 to 68 are prefixed by 20 and 69 to 99 are prefixed by 19.
* **%Y**: Year with century.
* **%H**: Hours as a decimal number (00-23).
* **%I**: Hours as a decimal number (01-12).
* **%p**: AM/PM indicator in the current locale.
* **%M**: Minute as a decimal number (00-59).
* **%S**: Second as a decimal number (00-61), allowing for leap seconds.
* **%a**: Abbreviated weekday name in the current locale (e.g., "Mon").
* **%A**: Full weekday name in the current locale (e.g., "Monday").
* **%j**: Day of the year as a decimal number (001-366).
* **%U**: Week number of the year (00-53) with Sunday as the first day of the week.
* **%W**: Week number of the year (00-53) with Monday as the first day of the week.

# <center> 📌ggplot2()</center>

Creates a variety of data viz by applying different visual properties to the data variables in R

**Benefits:**

* create different types of plots
* Customize look and feels of plot
* create high quality visuals
* combine data manipulation and visualisation using `%>%` operator

**Features most commonly used:**

* **aesthetics** - visual property of an object in your plot<br>
     ex: size, shape, color of data points
     
* **geoms** - geometric object used to represent data<br>
     ex: line graph / bar / scatter
     
* **facets** - provide subsets of data<br>
     ex: separate plots
     
* **labels and annotations** - customize plots description<br>
     ex: filter, legends etc.
     
**Commonly used Geom functions:**
1. geom_point() --> scatter plot
2. geom_bar() --> bar plot
3. geom_line() --> line graph
4. geom_smooth() --> shows trendline
5. geom_jitter() --> allow or provide clarity to overlapping data points
6. geom_histogram() --> histogram plot

```r
install.packages("palmerpenguins")
library(palmerpenguins)
data(penguins)
head(penguins)
```

**geom_smooth()**
Smoothing enables the detection of data trend even when you can't easily notice a trend from plotted data points.

types of smoothing:

- loess smoothing -> process is best for plots with less than or equal to 1000 points.
- Gam smoothing -> process is best for plots with greathan 1000 points.

```r
ggplot(data=penguins,aes(x = flipper_length_mm,y = body_mass_g)) + geom_point() + geom_smooth()
```

**Facet functions:**
- facet_wrap() -> Arranges plots in a one-dimensional ribbon. creates separate plots for each level of the variable.
- facet_grid() -> Arranges plots in a two-dimensional grid. creates subplots

```r
ggplot(data = penguins,aes(x=flipper_length_mm,y=body_mass_g))+ geom_point()+geom_smooth()+facet_wrap(~species)
```

**`~` Tidle operator:**

used to define relationship between dependent variable and independent variable in statistical model formula.

**The ~ variable part tells ggplot2 to create separate panels for each unique value of variable.**

- If you pass a single variable after the tilde, ggplot2 will create one panel for each unique value of that variable.

- Although facet_wrap is typically used with a single variable, it can handle interactions of multiple variables by using the interaction function interaction(variable1, variable2) within facet_wrap.

```r
ggplot(data = penguins,aes(x=flipper_length_mm,y=body_mass_g))+ geom_point()+geom_smooth()+facet_wrap(~interaction(species,sex))
```

**facet_grid(variable1 ~ variable2)** 
creates a grid of scatter plots with variable1 defining the rows and variable2 defining the columns.

```r
ggplot(data = penguins,aes(x=flipper_length_mm,y=body_mass_g))+ geom_point()+geom_smooth()+facet_grid(sex~island)
```

used to label title,subtitle,caption,x-axis,y-axis,tag, color of legends.

## qpvt() of pivottabler package

Summarizing function
`qpvt(data, rows, columns, summaries)` 

Parameters:

- data: The dataframe containing the data.
- rows: Variable(s) to be used for rows in the pivot table.
- columns: Variable(s) to be used for columns in the pivot table.
- summaries: A named vector specifying the summary statistics to be calculated.

**Summaries available:**
**min(), max(), mean(), median(), Q1 = quantile(prob=0.25,type=1), Q3 = quantile(prob = 0.75,type = 1), sd(),var(),n(),length()**
